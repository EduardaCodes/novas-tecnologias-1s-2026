### Operadores Aritméticos

```python
A = 2
B = 3

soma = A + B
subtracao = A - B
multiplicacao = A * B
divisao = A / B
potenciacao = A ** B
modulo = A % B

print(soma)
print(subtracao)
print(multiplicacao)
print(divisao)
print(potenciacao)
print(modulo)
```

### Estruturas Condicionais

```python
idade = int(input('Digite a idade: '))
possui_habilitacao = True

if idade >= 18 & possui_habilitacao:
    print('Possui Habilitação.')
elif idade >= 18:
    print('Sim')
    print('É maior de idade.')
else: 
    print('Não')
    print('Não é maior de idade.')
```

### Estruturas de Repetição

```python
for i in range(5):
    print(i)

cont = 0
while cont < 5:
    print(cont)
    cont += 1
```

```python
%%time
import math
n = 56
raiz_n = int(math.sqrt(n))
eh_primo = True
if n % 2 == 0:
    eh_primo = False
for i in range(3, raiz_n, 2):
    if n % i == 0:
        eh_primo = False

if eh_primo:
    print('É primo')
else:
    print('Não é primo')
```

### Estruturas de Dados

```python
precos_carros = [15000, 23000, 120000]  # Cria lista

# Adicionar outro elemento
precos_carros.append(60000)

print(precos_carros)

# Imprime um elemento específico
print(precos_carros[2])

# Remover um elemento
precos_carros.remove(120000)
print(precos_carros[2])
```

```python
carro1 = {
    "modelo" : "Monza GLS",
    "ano" : 1994,
    "preco" : 15000
}
carro2 = {
    "modelo" : "kwid",
    "ano" : 2026,
    "preco" : 60000
}

print(carro1)

#imprime chave específica
print(carro2["ano"])
```

```python
def calcular_parcela(valor_carro, qtd_parcelas, taxa):
    return ((valor_carro + (valor_carro*taxa))/qtd_parcelas)+10

taxa_financiamento = 0.1015
qtd_parcelas = 48
#valor_parcela ?

valor_parcela_carro1 = calcular_parcela(carro1['preco'], 12, taxa_financiamento)
valor_parcela_carro2 = calcular_parcela(carro2['preco'], 60, taxa_financiamento)

print(valor_parcela_carro1)
print(valor_parcela_carro2)
```
